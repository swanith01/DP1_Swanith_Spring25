```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
M = 1.989e30  # Example mass, 1 Solar mass (kg)
B = 1e13  # Arbitrary characteristic length scale (m)

# Compute A from A^3 = GMB^2
A = (G * M * B**2) ** (1/3)
```

```python
def overdensity(theta):
    return 9/2 * (theta - np.sin(theta))**2/(1 - np.cos(theta))**3 -1

# Define the linear approximation (Eq. 5.96)
def delta_linear(theta):
    return (3/20) * theta**2

theta_vals = np.linspace(0.01, 5, 100)  # More points for a smooth curve
```

```python
# Plot overdensity with log scale
fig, ax = plt.subplots()

#The Non virialised Friedmann evolution
ax.plot(theta_vals, overdensity(theta_vals), label=r'$\delta(\theta)$', color='blue')
# Add linear overdensity to the plot from jeans instabiltiy
ax.plot(theta_vals, delta_linear(theta_vals), linestyle='dashed', color='black', label=r'$\delta_{\mathrm{lin}}(\theta)$')
# Plot the virialized overdensity (constant after collapse)
# Add a horizontal line at the virialized theta
plt.axhline(178, linestyle="-", color="purple", label=r"$R_{\mathrm{vir}} = A$")


ax.set_yscale("log")  # Log scale for overdensity

# Labels and Title
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'Overdensity $\delta$')
ax.set_title("Overdensity Evolution in Spherical Top Hat Model")

# Add vertical lines for turnaround and collapse
ax.axvline(np.pi, linestyle='dashed', color='gray', label="Turnaround ($\theta = \pi$)")
ax.axvline(2*np.pi, linestyle='dotted', color='red', label="Collapse ($\theta = 2\pi$)")
ax.axvline(3*np.pi/2, linestyle='dotted', color='purple', label="Collapse ($\theta = 2\pi$)")

ax.legend()

plt.show()
```

```python
# Define theta values
theta_v = np.linspace(0, 2*np.pi, 500)

# Compute R(theta) = A(1 - cos(theta))
R_theta = 2*A * (1 - np.cos(theta_v))
#R i linear Regime
R_lin = 4*A*(theta_v -np.sin(theta_v))**(2/3)


# Plot the function
plt.figure(figsize=(8, 5))
plt.plot(theta_v, R_theta, label=r"$R(\theta) = A(1 - \cos\theta)$")
#Plot the linear too
plt.plot(theta_v, R_lin, label=r"$R_{\mathrm{lin}}(\theta)$")  
#Virialisation
# Add a horizontal line at the virialized radius R_vir
plt.axhline(A*2, linestyle="-", color="purple", label=r"$R_{\mathrm{vir}} = A$")

plt.axvline(np.pi, linestyle="--", color="red", label=r"$\theta = \pi$ (Max R)")
plt.axvline(2*np.pi, linestyle="--", color="green", label=r"$\theta = 2\pi$ (Collapse)")
ax.axvline(3*np.pi/2, linestyle='dotted', color='purple', label="Collapse ($\theta = 2\pi$)")


plt.xlabel(r"$\theta$")
plt.ylabel(r"$R(\theta)$ (meters)")
plt.title("Evolution of R() in the Spherical Collapse Model")
plt.legend()
plt.show()
```
